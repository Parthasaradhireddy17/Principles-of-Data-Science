Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

Upload the data set into google colab and read the dataset.


```python
# Load the dataset
df = pd.read_csv('/content/StudentsPerformance.csv')
```

Bar chart is used for Gender Distribution

```python

# Visualization 1: Bar Chart for Gender Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.show()


```

Scatter Plot is used for Math vs. Reading Scores

```python
# Visualization 2: Scatter Plot for Math vs. Reading Scores
plt.figure(figsize=(8, 6))
sns.scatterplot(x='math score', y='reading score', data=df)
plt.title('Math vs. Reading Scores')
plt.show()

```

Box Plot is used for Test Preparation Course vs. Writing Scores

```python
# Visualization 3: Box Plot for Test Preparation Course vs. Writing Scores
plt.figure(figsize=(8, 6))
sns.boxplot(x='test preparation course', y='writing score', data=df)
plt.title('Test Preparation Course vs. Writing Scores')
plt.show()

```

Histogram is used for Parental Education Levels

```python
# Visualization 4: Histogram for Parental Education Levels
plt.figure(figsize=(8, 6))
sns.histplot(x='parental level of education', data=df, hue='gender', multiple='stack')
plt.title('Parental Education Levels')
plt.xticks(rotation=45)
plt.show()
```

Pie Chart is used for Lunch Types

```python
# Visualization 5: Pie Chart for Lunch Types
lunch_counts = df['lunch'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(lunch_counts, labels=lunch_counts.index, autopct='%1.1f%%', startangle=140, colors=['#66b3ff','#99ff99'])
plt.title('Distribution of Lunch Types')
plt.show()
```
